```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset.csv')
plt.figure(figsize=(8, 6))
plt.hist(df['Aggregate rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
rating_bins = [0, 2, 3, 4, 5]
rating_labels = ['0-2', '2-3', '3-4', '4-5']
df['Aggregate rating'] = pd.cut(df['Aggregate rating'], bins=rating_bins, labels=rating_labels, include_lowest=True)
most_common_rating_range = df['Aggregate rating'].value_counts().idxmax()
average_votes = df['Votes'].mean()
```

```python
print(f"The most common rating range is: {most_common_rating_range}")
print(f"Average number of votes received by restaurants: {average_votes:.2f}")
```
